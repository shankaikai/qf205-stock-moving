```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("./CSV/AAPL.csv")

y = df["Adj Close"]

x = df["Date"]

plt.plot(x, y)
plt.legend(('Close'), loc='upper right')
plt.set_title('Stock price against date')
plt.set_xlabel('Dates')
plt.set_y_label('Prices')
plt.show()
```
